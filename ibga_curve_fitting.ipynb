```python

def hello(msg):
    print ("Your message is "+msg+".")

def c_fitting(data):
    import pandas as pd
    import numpy as np
    from scipy.optimize import curve_fit
    
    # Define the polynomial function
    def polynomial_func(x, *coefficients):
        return np.polyval(coefficients, x)
    
    # Read the CSV file
    df = pd.read_csv('IBGA Anomaly Individual all.csv')
    
    # Initialize a dictionary to store the fitting results
    fitting_results = {}
    
    # Group the data by 'ID'
    grouped = df.groupby('ID')
    
    # Perform curve fitting for each group
    for name, group in grouped:
        x = group['Hlimit'].values
        y = group['Defect Percent'].values
        
        # Initial guess for the coefficients (degree 2 polynomial)
        initial_guess = [1, 1, 1]
        
        # Perform the curve fitting
        try:
            popt, _ = curve_fit(polynomial_func, x, y, p0=initial_guess)
            fitting_results[name] = popt
        except RuntimeError:
            print(f"Curve fitting failed for ID {name}")
    
    # Print the fitting results
    for id, coefficients in fitting_results.items():
        print(f"ID {id}: Polynomial coefficients = {coefficients}")
```
